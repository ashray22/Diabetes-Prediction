```python
X = df.drop(['Outcome'], axis = 1)
```

```python
X
```

```python
# STANDARDIZING THE DATA
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X)
```

```python
X = sc.fit_transform(X)
```

```python
y = pd.DataFrame(df['Outcome'])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=100)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
# Model Training and HyperParameter Tuning ( LOGISTIC REGRESSION)

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
LR = LogisticRegression()
from sklearn.metrics import plot_confusion_matrix,classification_report,accuracy_score
param_grid = [    
    {'penalty' : ['l2', 'elasticnet'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [10, 100, 500,1000]
    }
]
from sklearn.model_selection import GridSearchCV
GSCV_lr = GridSearchCV(LR, param_grid = param_grid, cv = 4, verbose=True)
best_GSCV_lr = GSCV_lr.fit(X,y)
p1 = best_GSCV_lr.predict(X_test)
c1 = best_GSCV_lr.score(X,y)
a1= "{:.2f}%".format(100*c1)
best_GSCV_lr.best_estimator_
print ('Accuracy of LOGISTIC REGRESSION :', a1 )
plot_confusion_matrix(best_GSCV_lr,X_test,y_test,cmap='Blues')
plt.grid(False)
print("Classification report for Logitic Regression:\n ",classification_report(y_test,p1))
```

```python
best_GSCV_lr.best_params_
```

```python
# Model Training(RANDOM FOREST CLASSIFIER)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
clf = rf.fit(X_train,y_train)
y_pred_rf = clf.predict(X_test)
plot_confusion_matrix(clf,X_test,y_test,cmap='Blues')
plt.grid(False)
print("Classification report:\n ",classification_report(y_test,y_pred_rf))
print('Accuracy : ',metrics.accuracy_score(y_test,y_pred_rf))
```

```python
# Model Training And HyperParameter Tuning (RANDOM FOREST CLASSIFIER)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
GSCV_rf = GridSearchCV(rf,param_grid = param_grid, cv = 4, verbose = 2)

```

```python
best_GSCV_rf = GSCV_rf.fit(X,y)
```

```python
p2 = best_GSCV_rf.predict(X_test)
c2 = best_GSCV_rf.score(X,y)
a2= "{:.2f}%".format(100*c2)
best_GSCV_rf.best_estimator_
print ('Accuracy of Random Forest :', a2 )
plot_confusion_matrix(best_GSCV_rf,X_test,y_test,cmap='Blues')
plt.grid(False)
print("Classification report for Random Forest:\n ",classification_report(y_test,p2))
```

```python
best_GSCV_rf.best_params_
```

```python
# Model Training And HyperParameter Tuning (DECISION TREE CLASSIFIER)
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
param_grids = {
    'max_depth': [2, 3, 5, 10, 20,35,60],
    'min_samples_leaf': [5, 10, 20, 50, 100,200],
    'criterion': ["gini", "entropy"]
}
GSCV_dt = GridSearchCV(estimator=dt, param_grid=param_grids, cv=4, n_jobs=-1, verbose=1)
```

```python
best_GSCV_dt = GSCV_dt.fit(X,y)
```

```python
p3 = best_GSCV_dt.predict(X_test)
c3 = best_GSCV_dt.score(X,y)
a3= "{:.2f}%".format(100*c3)
best_GSCV_dt.best_estimator_
print ('Accuracy of Decision tree :', a3 )
plot_confusion_matrix(best_GSCV_dt,X_test,y_test,cmap='Blues')
plt.grid(False)
print("Classification report for Decision Tree Classifier :\n ",classification_report(y_test,p3))
```

```python
best_GSCV_dt.best_params_
```
